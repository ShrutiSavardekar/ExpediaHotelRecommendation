```python
import matplotlib.pyplot as plt
 
# x-coordinates of left sides of bars
left = [1, 2, 3, 4, 5]
 
# heights of bars
height = [0.9613, 0.9659, 0.2807, 0.2700, 0.2880]
 
# labels for bars
tick_label = ['User-based CF', 'Item-based CF', 'Wide and Deep', 'DeepFM', 'XDeepFM']
 
# plotting a bar chart
plt.bar(left, height, tick_label = tick_label,
        width = 0.8, color = ['green'])
 
# naming the x-axis
plt.xlabel('Models')
# naming the y-axis
plt.ylabel('RMSE values')
# plot title
plt.title('Comparing RMSE values')
 
# function to show the plot
plt.show()
```
